```python
makanan = set(["bengbeng", "burger", "bengbeng"])
print("sebelum ditambahkan : ", makanan)
makanan.add ("keju")
makanan.add ("pizzaaa")
print("sesudah ditambahkan : ", makanan)

# command "update" bisa menambahkan banyak data menggunakan "([])"
# pada saat ditambahkan UPDATE maka :
makanan.update (["kaefci", "mekudi"])
print("setelah diUPDATE :" , makanan)
```

```python
def printElements(s,ingpoo):
    print(ingpoo)
    if len(s) == 0:
        print("himpunan kosong")
    for i in s: # untuk index dalam s
        print(i) # panggilan kpd index

# masukkan data untuk "s" dan panggil print
s = set([12, 32, 43, 65])
printElements(s,"sebelum")

# command "add" itu hanya bisa menambahkan satu data menggunakan "()"
s.add(88)
s.add(77)
printElements(s,"sesudah panggilan add():")

# command "update" bisa menambahkan banyak data menggunakan "([])"
s.update([400,700])
printElements(s,"setelah dipudate ([]):")
```

```python
x = set([2,4,5,1])
y = set([2,7,4,8])

# untuk menggabungkan data x dan y
union = x | y
print("hasil dari union :", union)

# untuk data yang sama pada x & y
inter_section = x & y
print("hasil dari section:",inter_section)

# ada di set awal tapi tidak ada di set akhir
difference = x - y
print("hasil dari difference:", difference)

# data yang berbeda pada x & y
symetric_difference = x ^ y
print("hasil dari symetric_difference:", symetric_difference)
```
