```python
#Type Numerik
#Type Integer(int)
a = 1
b = -17
c = 100

print(type(a))
print(type(b))
print(type(c))
```

```python
#Type Float
x = 3.14
y = -0.175
z = 2.05

print(type(x))
print(type(y))
print(type(z))
```

```python
#Type Complex
s1 = 3 + 2j
s2 = -2.5j

print(type(s1))
print(s1.real) #menampilkan bilangan real
print(s1.imag) #menampilkan bilangan imajiner
print(type(s2))
print(s2.real) #menampilkan bilangan real
print(s2.imag) #menampilkan bilangan imajiner
```

```python
#Konversi Tipe Data Numerik
#Konversi ke integer(int)
x = int(3.14)
print(x)

#Konversi ke float
b = float(-17)
print(b)

#Konversi ke complex
y = complex(-0.175)
print(y)
```

```python
#Operasi Aritmatika
a = 15 
b = 6
print(a+b) #penjumlahan
print(a-b) #pengurangan
print(a*b) #perkalian
print(a/b) #pembagian
print(a//b) #Untuk pembagian bilangan bulat
print(a%b) #modulus
print(a**b) #eksponen
```

```python
#Modul match
import math
a = 4
b = 2
c = 0
print(math.sqrt(a)) #Akar variabel a = 4
print(math.pi*(b)) #Perkalian variabel b = 2 dengan pi = 3.14
print(math.cos(c)) #Cosinus variabel c = 0
```

```python
#Modul cmath
import cmath
z= 1 + 4j
print(cmath.sqrt(z)) #akar variabel z
print(cmath.sin(z)) #sinus variabel z
```
