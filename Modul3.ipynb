```python
#VAriabel dictionary
dic = {'satu':5, 'dua':10}

print("satu:",dic['satu'])
print("dua:",dic['dua'])

print("dic['satu'] * dic['dua']:" , (dic['satu'] *  dic['dua']))

#Operasi mtematis
print("satu * dua:" , (dic['satu'] *  dic['dua']))
print("satu + dua:" , (dic['satu'] +  dic['dua']))
print("satu / dua:" , (dic['satu'] /  dic['dua']))

#Menambah elemen
dic['tiga'] = 165
print(dic)

#Mengubah Elemen
dic['dua'] = 30
print(dic)

#Menghapus Elemen
del dic['dua']
print(dic)

#SOrting 
D = {'a':1, 'b':2, 'c':3}
print(D)
Ks = list(D.Keys())
print(Ks)
Ks.sort()
print(Ks)

for key in Ks:
    print(key, '=', D[key])
```
